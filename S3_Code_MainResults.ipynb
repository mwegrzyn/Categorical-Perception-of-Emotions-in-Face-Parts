## Get List of previously saved df's (pickle)

```python
def get_list(where, what):
    #os.chdir(where)
    fileList = []
    for fileName in os.listdir(where):
        if fnmatch.fnmatch(fileName, what):
            fileList.append(where+fileName)
    return fileList
```

## Inferential Statistics

```python
def inf_stats(df,conds,my_alpha):
    for c1 in conds:
        for c2 in conds:
            if c1 != c2:
                print c1,c2
                for m in df.columns.levels[1]:
                    t,p = stats.ttest_rel(df[c1][m],df[c2][m])
                    T,np = stats.wilcoxon(df[c1][m],df[c2][m])
                    if p < my_alpha and np < my_alpha:
                        sig = '*'
                    else:
                        sig = 'n.s.'
                    
                    print m,'\tt:',round(t,2),'p:',round(p,3),'\tT:',round(T,2),'np:',round(np,3), sig
```

Pairwise differences between all conditions for all morphing grades:

```python
inf_stats(exp1Table,['whole','mouth','eyes'],0.05)
```

Pairwise differences of the differences 'whole-eyes' and 'whole-mouth':<br>
This analysis asks whether the eyes are more similar to the whole face than the mouth is.

```python
diff_whole_eyes = exp1Table['whole']-exp1Table['eyes']
diff_whole_mouth = exp1Table['whole']-exp1Table['mouth']

for m in exp1Table.columns.levels[1]:
    t,p = stats.ttest_rel(diff_whole_eyes[m],diff_whole_mouth[m])
    W,wp = stats.wilcoxon(diff_whole_eyes[m],diff_whole_mouth[m])
    if p < 0.05 and wp < 0.05:
        sig = '*'
    else:
        sig = 'n.s.'
                    
    print m,'\tt:',round(t,2),'p:',round(p,3),'\tT:',round(W,2),'np:',round(wp,3), sig
```

### Comparison of Bias between Composite Face Conditions

The difference of an angry or fearful to-be-ignored half is compared between upper and lower halves.<br>
This analysis asks whether the biasing influence of the to-be-ignored half is stronger for the eyes over the mouth or for the mouth over the eyes.

```python
low_diff = exp2Table['angerLOW']-exp2Table['fearLOW']
up_diff = exp2Table['angerUP']-exp2Table['fearUP']
for m in ['m00','m01','m02','m03','m04','m05','m06','m07','m08','m09','m10']:
    t,p = stats.ttest_rel(up_diff[m],low_diff[m])
    W,wp = stats.wilcoxon(up_diff[m],low_diff[m])
    if p < 0.001 and wp < 0.001:
        sig = '*'
    else:
        sig = ''
    print m,'\tt:',round(t,2),'p:',round(p,3),'\tW:',round(W,2),'wp:',round(wp,3), sig
    
```
